```python
import pandas as pd
import json
```

```python
df = pd.read_json('dataset/full_train_dataset.json',orient='records')
df
```

```python
df['tag_string_no_meta'] = df.apply(lambda row: ', '.join([tag for tag in row['tag_string'].split() if tag not in row['tag_string_meta'].split()]), axis=1)
df['tag_string_no_meta'][0]
```

```python
from utils.similar_tag import load_dict
tag_dict = load_dict()

df['tag_filtered'] = df['tag_string_no_meta'].apply(lambda row: ', '.join([x for x in row.split(', ') if x in tag_dict]))
df['tag_filtered'][0]
```

```python
train_data_df = df[['caption_string', 'tag_filtered']]
train_data_df.columns = ['caption_string', 'tag_string']
result = train_data_df.to_json(orient='records')
parsed = json.loads(result)
with open('train_data_v4.json', 'w') as outfile:
    json.dump(parsed, outfile, indent=4)
```

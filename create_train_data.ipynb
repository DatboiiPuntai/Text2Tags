```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import json
```

```python
df = pd.read_json('dataset/data_captioned.json',orient='records')
df
```

```python
df['tag_string_no_meta'] = df.apply(lambda row: ', '.join([tag for tag in row['tag_string'].split() if tag not in row['tag_string_meta'].split()]), axis=1)
df['tag_string_no_meta'][0]
```

```python
df = df[df['prompt_format_id'] != 0].reset_index(drop=True)
df
```

```python
train_data_df = df[['caption_string', 'tag_string_no_meta']]
train_data_df.columns = ['caption_string', 'tag_string']
result = train_data_df.to_json(orient='records')
parsed = json.loads(result)
with open('train_data_v2.json', 'w') as outfile:
    json.dump(parsed, outfile, indent=4)
```

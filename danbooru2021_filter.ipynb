```python
import pandas as pd
```

```python
df = pd.read_json('danbooru2021\posts000000000010.json', orient='records', lines=True)
df
```

```python
df.info()
```

```python
df_reduced = df.iloc[:100000]
df_reduced
# df_reduced['timestamp'] = pd.to_datetime(df_reduced['updated_at'])
```

```python
def filter_row(row, min_tags=15, min_year=2018):
    if row['is_deleted']:
        return False
    if row['is_banned']:
        return False
    if row['created_at'].year < min_year:
        return False
    if len(row['tag_string'].split(' ')) <= min_tags:
        return False
    return True
```

```python
df_filtered = df_reduced[df_reduced.apply(filter_row,axis=1)]
df_filtered
```

```python
df_filtered_sampled = df_filtered.iloc[:150].reset_index(drop=True)
df_filtered_sampled
```

```python
df_filtered_sampled['tag_count'].hist()
```

```python
df_filtered['tag_count'].hist()
```

```python
json_str = df_filtered_sampled.to_json(orient='records')
with open('test_data.json', 'w') as f:
    f.write(json_str)
```
